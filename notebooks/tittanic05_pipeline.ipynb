```python
df = pd.read_csv("https://raw.githubusercontent.com/ananyaghorpade29/titanic/refs/heads/main/data/Titanic-Dataset.csv")
df.head()
```

```python
x= df.drop(columns=["Name","PassengerId","Ticket","Survived"])
y = df["Survived"]

categorical_col = ["Sex","Embarked"]
numerical_col = ["Age","Pclass","Fare","SibSp","Parch"]

numeric_pipeline= Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_pipeline= Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(drop="first"))
])

preprocessor= ColumnTransformer([
    ("num",numeric_pipeline, numerical_col),
    ("cat", categorical_pipeline, categorical_col)
])
```
